Normalizing certain dataset columns

```python
from sklearn.preprocessing import RobustScaler,StandardScaler,MinMaxScaler
robustScaler = RobustScaler()
standardScaler = StandardScaler()
minMaxScaler = MinMaxScaler()
Amount = data_set['Amount'].to_numpy().reshape(-1,1)
Time = data_set['Time'].to_numpy().reshape(-1,1)
data_set['Amount'] = robustScaler.fit_transform(Amount)
data_set['Time'] = minMaxScaler.fit_transform(Time)
data_set.describe()
```

Displaying the normalized dataset in histograms

```python
data_set.hist(bins=30, figsize=(50,50))
```

```python
#Splitting the data to features and targets.
X = data_set.drop('Class', axis=1)
y = data_set['Class']
```

```python
from imblearn.over_sampling import SMOTE
#Resample the training dataset to be balanced.
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
from collections import Counter
print(Counter(y_resampled))
```

Using Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(X_train, y_train)
logistic_regression_model.score(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test, logistic_regression_model.predict(X_test), target_names=['Not Fraud', 'Fraud']))
print(confusion_matrix(y_test, logistic_regression_model.predict(X_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree_model = DecisionTreeClassifier( max_depth=10,
                                              min_samples_split=20,
                                              min_samples_leaf=10,
                                              class_weight="balanced",
                                              random_state=42,
                                              max_leaf_nodes=50,)
decision_tree_model.fit(X_train, y_train)
decision_tree_model.score(X_train, y_train)
```

```python
from sklearn import tree
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(decision_tree_model, 
                   feature_names=X_train.columns,  
                   class_names={0:'Not Fraud', 1:'Fraud'},
                   filled=True,
                  fontsize=12)
```

```python
print(classification_report(y_test, decision_tree_model.predict(X_test), target_names=['Not Fraud', 'Fraud']))
print(confusion_matrix(y_test, decision_tree_model.predict(X_test)))
print(Counter(y_test))
```
